```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.svm import OneClassSVM
from sklearn.metrics import classification_report, confusion_matrix
import tensorflow as tf
from tensorflow import keras



df = pd.read_excel("FinalDataset.xlsx", sheet_name="Dataset")

# 3. Select Numeric Features

numeric_cols = ["value", "gas", "gas_price", "receipt_gas_used", "block_number"]
X = df[numeric_cols].fillna(0)

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


# 4. One-Class SVM
svm = OneClassSVM(kernel="rbf", gamma="auto", nu=0.05)  # nu=anomaly proportion
svm.fit(X_scaled)

df["svm_pred"] = svm.predict(X_scaled)  # -1 = anomaly, 1 = normal

print("SVM Predictions:", df["svm_pred"].value_counts())

input_dim = X_scaled.shape[1]
input_layer = keras.layers.Input(shape=(input_dim,))
encoder = keras.layers.Dense(16, activation="relu")(input_layer)
bottleneck = keras.layers.Dense(8, activation="relu")(encoder)
decoder = keras.layers.Dense(16, activation="relu")(bottleneck)
output_layer = keras.layers.Dense(input_dim, activation="linear")(decoder)

autoencoder = keras.Model(inputs=input_layer, outputs=output_layer)
autoencoder.compile(optimizer="adam", loss="mse")

# Train Autoencoder
history = autoencoder.fit(
    X_scaled, X_scaled,
    epochs=20,
    batch_size=32,
    validation_split=0.2,
    verbose=1
)

# Reconstruction Errors
reconstructions = autoencoder.predict(X_scaled)
mse = np.mean(np.square(X_scaled - reconstructions), axis=1)
threshold = np.percentile(mse, 95)  # top 5% as anomalies
df["autoencoder_pred"] = (mse > threshold).astype(int)  # 1=anomaly, 0=normal

print("Autoencoder Predictions:", df["autoencoder_pred"].value_counts())


# 6. Visualization

plt.figure(figsize=(8,5))
plt.hist(mse, bins=50, color="blue", alpha=0.7)
plt.axvline(threshold, color="red", linestyle="--", label="Anomaly Threshold")
plt.xlabel("Reconstruction Error")
plt.ylabel("Count")
plt.title("Autoencoder Anomaly Detection")
plt.legend()
plt.show()
```
